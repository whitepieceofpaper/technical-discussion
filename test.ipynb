```python
# Function 1
def increment(x):
    return x + 1

def test_increment():
    assert increment(3) == 4
    assert increment(5) == 6, "Increment failed"
    
    #Catch failures with try/except
    try:
        assert increment(-2) == -1 
    except AssertionError:
        print("Incrementing negative number failed")
        
```

```python
test_increment()
```

```python
# Function 2
def square(num):
    return num ** 2


def test_square():
    assert square(2) == 4
    assert square(5) == 25
    assert square(-3) == 9
```

```python
try:
    test_square()
    print("Square Test passed!")
except AssertionError:
    print("Square Test failed")
```

```python
# Function 3
def is_even(num):
    return num % 2 == 0


def test_is_even():
    assert is_even(2) is True
    assert is_even(7) is False
    assert is_even(0) is True
```

```python
try:
    test_is_even()
    print("Even Test passed!")
except AssertionError:
    print("Even Test failed")
```
